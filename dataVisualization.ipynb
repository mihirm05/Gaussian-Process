```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
dfN = df[df.columns.difference(['Country', 'Status'])]
```

```python
df_clean = dfN[~((dfN < (Q1 - 1.5 * IQR)) |(dfN > (Q3 + 1.5 * IQR))).any(axis=1)]
```
